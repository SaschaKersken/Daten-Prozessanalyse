```python
import matplotlib.pyplot as plt
iris.groupby(['Classification']).plot.box()
```

```python
import seaborn as sns
sns.FacetGrid(iris,hue='Classification',height=5)\
.map(plt.scatter,'SepalLength','SepalWidth')\
.add_legend()
```

```python
sns.scatterplot(
    data=iris, x='SepalLength', y='SepalWidth',
    hue='Classification', style='Classification'
)
```

```python
sns.boxplot(x='Classification',y='PetalLength',data=iris)
```

```python
sns.pairplot(data=iris,kind='scatter',hue='Classification')
```

```python
sns.heatmap(iris.corr(), annot=True, cmap='viridis')
```
